```python
from os import walk as oswalk

PATH = r'.venv/' 

Nfiles=0
Ndirs=0

for root, dirnames,filenames in oswalk(PATH):
    Nfiles+=len(dirnames)
    Ndirs+=len(filenames)

print("Ndirs",Ndirs)
print("Nfiles",Nfiles)
```
